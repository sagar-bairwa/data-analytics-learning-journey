```python
### pie chart

labels=['A','B','C','D']
sizes=[30,20,40,10]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0,0,0.2,0) ##move out the 1st slice

##create apie chart
mpl.pie(sizes) #only compistions of pie chart out of 100% is compulsory
mpl.show()

mpl.pie(sizes,labels=labels,colors=colors)
mpl.show()

mpl.pie(sizes,labels=labels,colors=colors,explode=explode)
mpl.show()

mpl.pie(sizes,labels=labels,colors=colors,explode=explode,autopct="%1.1f%%")
mpl.show()

mpl.pie(sizes,labels=labels,colors=colors,explode=explode,autopct="%1.1f%%",shadow=True)
mpl.show()

mpl.pie(sizes,labels=labels,colors=colors,explode=explode,autopct="%1.1f%%",shadow=True, startangle=90)
mpl.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels  = ['A', 'B', 'C', 'D']
sizes   = [30, 20, 40, 10]
colors  = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0, 0.2, 0)

fig, axs = plt.subplots(2, 3, figsize=(15, 8))  # 2 rows, 3 cols = 6 plots
axs = axs.flatten()

configs = [
    dict(),
    dict(labels=labels, colors=colors),
    dict(labels=labels, colors=colors, explode=explode),
    dict(labels=labels, colors=colors, explode=explode, autopct="%1.1f%%"),
    dict(labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True),
    dict(labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True, startangle=90),
]

titles = [
    "Basic Pie",
    "With Labels & Colors",
    "With Explode",
    "With Autopct",
    "With Shadow",
    "Full Styled Pie",
]

for ax, config, title in zip(axs, configs, titles):
    ax.pie(sizes, **config)
    ax.set_title(title)

plt.tight_layout()
plt.show()
```
