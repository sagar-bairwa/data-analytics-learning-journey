Fill something at the place of null values.

```python
filled_df = df.fillna(0)
filled_df.head(18)
```

Fill mean value of the column where null values are present

```python
df['Value_fillna'] = df['Value'].fillna(df['Value'].mean())
df.head(18)
```

Change the data type of the columns

```python
df['Value_fillna'] = df['Value_fillna'].astype(int)
df.head()
```

Apply function on specific column

```python
def square(num):
    return num*num

df['new_value'] = df['Value_fillna'].apply(square)
df.head()
```

Grouped columns to get mean of each group

```python
grouped_mean = df.groupby('Category')['Value'].mean()
grouped_mean.head()
```

Grouped specofic columns to get desires output in ascending or descending order

```python
df.groupby('Product')['Sales'].sum().sort_values(ascending=True)
```

Aggregate multiple functions on grouped data

```python
df.groupby(['Category','Product'])['Value'].agg(['mean', 'sum', 'max', 'min'])
```

Two different DataFrames to apply merge operations on them

```python
import pandas as pd

customers = pd.DataFrame({
    "customer_id": [101, 102, 103, 104, 105],
    "name": ["Amit", "Priya", "Rahul", "Neha", "Karan"],
    "city": ["Delhi", "Mumbai", "Pune", "Delhi", "Bangalore"]
})

orders = pd.DataFrame({
    "order_id": [5001, 5002, 5003, 5004, 5005],
    "customer_id": [101, 103, 104, 106, 102],
    "product": ["Laptop", "Phone", "Tablet", "Camera", "Headphones"],
    "amount": [70000, 20000, 15000, 30000, 5000]
})
```

```python
customers
```

```python
orders
```

Inner merge operations

```python
pd.merge(customers,orders, on='customer_id', how='inner')
```

Outer merge operation

```python
pd.merge(customers,orders, on='customer_id', how='outer')
```

Left merge operation

```python
pd.merge(customers,orders, on='customer_id', how='left')
```

Right merge operation

```python
pd.merge(customers,orders, on='customer_id', how='right')
```

After that apply groupby on merge dataframe and apply agg function on it.

```python
merged_df = pd.merge(customers, orders, on="customer_id", how="inner")
merged_df.groupby(['name','product'])['amount'].sum()
```

```python
merged_df.groupby(['name', 'customer_id', 'product'])['amount'].sum()
```
