# Seaborn Library Functions

```python
import seaborn as sb 

data = sb.load_dataset('tips')
data
```

```python
sb.scatterplot(data=data , x='total_bill' , y='tip') #show spread of data

```

```python
sb.lineplot(data=data , x='size' , y='tip') # show trendline
```

```python
sb.boxplot(data=data, x='day', y='total_bill')
```

```python
data
```

```python
sb.boxplot(data=data, x='day' ,y='size')
```

```python
sb.violinplot(data=data, x='day', y='total_bill')
```

```python
data
```

```python
sb.violinplot(data=data, x='day', y='tip')
```

```python
sb.histplot(data['tip'] ,bins=10,)
```

```python
sb.histplot(data['total_bill'], bins=12, kde=True)
```

```python
sb.kdeplot(data['total_bill'])
```

```python
sb.kdeplot(data['tip'], fill=True)
```

```python
sb.pairplot(data=data)
```

```python
data
```

```python
corr = data[['total_bill', 'tip', 'size']].corr()
corr
```

```python
sb.heatmap(corr, annot=True)
```

```python
sb.heatmap(corr, annot=True, cmap='coolwarm')
```
